```
"""Ejemplo de manejo de excepciones"""

while True:
    # Intentar convertir y dividir valores
    try:
        numero1 = int(input('Ingresa el numerador: '))
        numero2 = int(input('Ingresa el denominador: '))
        resul = numero1 / numero2
    except ValueError:  # Intentó convertir un no numerico en entero
        print('Debes ingresar dos números\n')
    except ZeroDivisionError:  # El denominador fue cero
        print('División entre cero\n')
    else:  # Se ejecuta en el caso de que no hubo excepciones 
        print(f'{numero1:.3f} / {numero2:.3f} = {resul:.3f}')
        break  # termina el ciclo while
```
