**Para modelos pre-entrenados**: Usa las siguientes celdas para copiar los modelos pre-entrenados a la carpeta .keras/models Esto sirve para evitar descargar los modelos pre-entrenados. Aunque lo mejor es descargarlos porque los modelos preentrenados proporcionados por el curso pueden estar desactualizados. 

```
import shutil
import os

source = './00_modelos_preentrenados/models'
dest = '/root/.keras/models'

if not os.path.exists(dest):
  os.mkdir(dest) # si no existe la carpeta /root/.keras/models
files = os.listdir(source)

for f in files:
    print(f)   
    src = os.path.join(source, f)
    dst = os.path.join(dest, f)
    shutil.copyfile(src, dst)
```
