```
def suma_resta(dato_1, dato_2):
    """Se calcula la suma y resta de 
    dos datos dados"""
    suma = dato_1 + dato_2
    resta = dato_1 - dato_2
    return suma, resta
```

```
resul = suma_resta(4, 5)
print(resul) 
```
